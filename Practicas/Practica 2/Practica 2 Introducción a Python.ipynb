```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una lista de años en el rango de 1629 a 1710
año = list(range(1629, 1711))

# Crear datos ficticios de nacimientos de hombres y mujeres
np.random.seed(0)  # Para reproducibilidad
niños = np.random.randint(100, 1000, len(año))
niñas = np.random.randint(100, 1000, len(año))

# 1. Crear un histograma de los nacimientos de hombres y mujeres
plt.hist([niños, niñas], bins=20, label=['Niños', 'Niñas'])
plt.xlabel('Número de Nacimientos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Nacimientos de Niños y Niñas en Londres (1629-1710)')
plt.legend()

# 2. Agregar una cuarta columna que muestre el total de nacimientos
total_nacimientos = niños + niñas

# 3. Agregar una quinta columna que muestre la proporción de hombres del total de nacimientos
prop_hombres = niños / total_nacimientos

# 4. Agregar una sexta columna que muestre la proporción de mujeres del total de nacimientos
prop_mujeres = niñas / total_nacimientos

# 5. Agregar una sexta columna con la comparación prop_hombre > prop_mujeres
comparacion = prop_hombres > prop_mujeres

# Imprimir las listas
print("Total de nacimientos:", total_nacimientos)
print("Proporción de hombres:", prop_hombres)
print("Proporción de mujeres:", prop_mujeres)
print("Comparación:", comparacion)

# Mostrar el histograma
plt.show()
```
