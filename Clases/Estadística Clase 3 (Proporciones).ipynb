```python
import random
```

### Población

```python
unos = [1]
no_def = (unos * 9900) # Crea una lista de 9900 no defectuosos (1)
print(len(no_def))
```

```python
ceros = [0]
defect = (ceros * 100) # Crea una lista de 100 defectuosos (0)
print(len(defect))
```

```python
poblacion = (defect + no_def) # Concatena las listas
random.shuffle(poblacion)
print(len(poblacion))
```

### 1 Muestra

```python
muestra1 = random.sample(poblacion, 1000) # Extraer una muestra de 1000 elementos
print(len(muestra1))
```

```python
unos_m1 = muestra1.count(1)
print("No defectuosos:", unos_m1)

ceros_m1 = muestra1.count(0)
print("Defectuosos:", ceros_m1)
```

```python
proporcion_def1 = ceros_m1 / len(muestra1)
print("Proporción defectuosos:", proporcion_def1)
```

### 5 muestras

```python
muestras = []
for i in range(5):
    muestra = random.sample(poblacion, 1000)
    muestras.append(muestra)

# Imprimir las muestras aleatorias
for i, muestra in enumerate(muestras):
    print("Muestra", i+1, ":", muestra)
```
