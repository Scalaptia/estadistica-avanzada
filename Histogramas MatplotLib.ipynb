```python
# Importar librerias para estadísticas y graficar
import numpy as np
import matplotlib.pyplot as plt

# Definir parámetros
tamaño_muestra = 10000 
confianza = 0.99
media = 2.6  # Valor esperado de la población
desviacion = 0.3  # Desviación estándar de la población
z = 1.96  # Valor crítico para el nivel de confianza del 99%

# Calcular margen de error
margen_err = z * (desviacion / np.sqrt(tamaño_muestra)) # Fórmula para calcular márgen de error
men_confianza = media - margen_err  # Límite inferior del intervalo de confianza
may_confianza = media + margen_err  # Límite superior del intervalo de confianza

# Imprimir resultados
print(f"Tamaño Muestra: {tamaño_muestra}")
print(f"Media Muestral: {media}")
print(f"Desviación Muestral: {desviacion}")
print(f"Intervalo de confianza (99%): {men_confianza} - {may_confianza}")
print(f"Margen de error: {margen_err * 2}")

# Generar datos de muestra con distribución normal
datos_muestra = np.random.normal(media, desviacion, tamaño_muestra)

# Crear un histograma para visualizar la distribución
plt.hist(datos_muestra)
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia de valores')

# Líneas que representan el intervalo de confianza
plt.axvline(men_confianza, color='red', linestyle='dashed')
plt.axvline(may_confianza, color='red', linestyle='dashed')
```
