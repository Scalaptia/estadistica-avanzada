```python
# Importar librerias para estadísticas y graficar
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import random

# Definir parámetros
tamaño_muestra = 10000
confianza = 0.99
calificaciones = list(range(0, 101))
muestra = random.choices(calificaciones, k = tamaño_muestra)

med_calif = np.mean(muestra)
des_calif = np.std(muestra)
z = 2.575

# Calcular margen de error
margen_err = z * (des_calif / np.sqrt(tamaño_muestra)) # Fórmula para margen de error.
men_confianza = med_calif - margen_err
may_confianza = med_calif + margen_err

# Imprimir resultados
print(f"Tamaño Muestra: {tamaño_muestra}")
print(f"Media Muestral: {med_calif}")
print(f"Desviación Muestral: {des_calif}")
print(f"Intervalo de confianza (99%): {men_confianza} - {may_confianza}")
print(f"Margen de error: {margen_err * 2}")

# Generar datos de muestra con distribución normal
datos_muestra = muestra
# datos_muestra = np.random.normal(med_calif, des_calif, tamaño_muestra)

# Crear un histograma para visualizar la distribución
plt.hist(datos_muestra, bins = 25, edgecolor='black')
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia de valores')

# Líneas que representan el intervalo de confianza
plt.axvline(men_confianza, color='red', linestyle='dashed')
plt.axvline(may_confianza, color='red', linestyle='dashed')
```
